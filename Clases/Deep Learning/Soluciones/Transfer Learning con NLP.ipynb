# Que es transfer Learning?

Transfer Learning, o conocimiento transferido, no es más que ser capaces de usar el conocimiento de una tarea aplicada a otra tarea.

![Transfer Learning](http://ruder.io/content/images/2017/03/transfer_learning_setup.png "Transfer Learning")

Cuando queremos usar Transfer learning?

2 escenarios:

*   Tareas semanticamente similares.
    *   Sentiment Analysis 
*   Tareas semanticamente distintas pero podemos transferir cierto conocimiento aka parametros. Por ejemplo, tenemos 2 redes similares, pero una tarea tiene 5 classes y la otra 10.

El escenario más común y visto es el siguiente. Queremos resolver una tarea t, para la que tenemos muy pocos datos. Pero, tenemos una tarea s, con datos semanticamente "similares". Entonces montamos un modelo para la tarea s, entrenamos, y una vez entrenamos, dejamos la red toda "congelada", y entrenamos la ultima capa con los datos de la tarea t. Se hace muy a menudo en Computer Vision, pero tambien es habitual verlo en NLP.

![Transfer Learning](https://image.slidesharecdn.com/producttank-ai-braincreators-170412101012/95/deep-neural-networks-embeddings-and-transfer-learning-76-638.jpg?cb=1511526967 "Transfer Learning")

En el caso de NLP, los word embeddings, es el ejemplo más claro de transfer learning. Entrenamos con datasets enormes los word embeddings, y luego usamos estos word embeddings pre-entrenados en cualquier otra tarea. Es muy tipico usar estos word embeddings y hay varios pre-entrenados, como GloVe (stanford), fasttext (facebook), word2vec (Mikolov -- google).
